**pattern-3**

```python
range(start,stop,step)
step value
-Postive step value
-if step value is positive: increment
-last = stop-1

-negative step value
-if step value is negative: decrement
-last = stop+1
```

```python
#case-1:
for i in range(0,20,2):
    print(i,end=' ')
```

```python
# case-2:
for i in range(2,20,-2):
 print(i,end=' ')
# start=2
# step= -2 negative direction
# last= stop+1 : 20+1 =21
```

```python
#case-3:
for i in range(20,2,-2):
    print(i,end= ' ')
# start=20
# step= -2 negative
# last= stop+1= 2+1 =3
# 20 18 16 14 12 10 8 6 4 
```

```python
# case-4:
for i in range(-20,2,-2): 
 print(i,end=' ')
# start= -20
# step= -2 neg
# last= stop+1 = 2+1=3
```
